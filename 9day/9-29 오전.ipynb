```python
import pandas as pd

exam_data = {'이름' : [ '서준', '우현', '인아'],
             '수학' : [ 90, 80, 70],
             '영어' : [ 98, 89, 95],
             '음악' : [ 85, 95, 100],
             '체육' : [ 100, 90, 90]}
df = pd.DataFrame(exam_data)

# '이름' 열을 새로운 인덱스로 지정하고, df 객체에 변경사항 반영
df.set_index('이름', inplace=True)
print(df)
```

```python
# 서준의 수학 점수를 100으로 변경하세요
# 우현과 인아의 영어와 수학점수를 출력하세요 ( iloc와 범위 , loc와 배열 사용)
# 인아의 모든 점수를 100으로 수정하세요 
# 서준의 모든 점수 출력
# 수학과 음악 점수만 출력

df.loc['서준','수학'] = 100
print(df.loc[['우현','인아'],['영어','수학']])
print()
df.iloc[2, 0:] = 100
print(df.loc['인아', '수학':])
print()
print(df.iloc[0])
print()
print(df.iloc[0:, [0, 2]])

df
```

```python
#reindex :index를 재배치

import pandas as pd

# 딕셔서리를 정의
dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}

# 딕셔서리를 데이터프레임으로 변환. 인덱스를 [r0, r1, r2]로 지정
df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
print(df)
print('\n')
```

```python
new_index = ['r0', 'r1', 'r2', 'r3', 'r4', 'r5']
ndf = df.reindex(new_index) # .reindex(객체), fill_value 로 초기값 설정 가능.
print(ndf)
```

```python
import pandas as pd
dict_data = {"c0":[1,2,3], "c1":[4,5,6], "c2":[7,8,9], "c3":[10,11,12], "c4":[13,14,15]}

df= pd.DataFrame(dict_data, index=['r0','r1','r2'])
print(df)
print('\n')

ndf = df.reset_index()
print(ndf)
```

```python
#시리즈와 숫자 연산
student1 = pd.Series({'국어':100, '영어':80, '수학':90})
student2 = pd.Series({'국어':90, '영어':85, '수학':95})
print(student1)
print('\n')

student1/200
(student1+student2)/2 #시리즈 연산자  (+ - / *) 가능 또는 숫자

student1.add(student2, fill_value=0) # add, sub, mul, div(자료가 없으면 원하는 값으로 대체)

#연산 식에 NaN이 포함 되면 결과는 NaN -> 옵션으로 fill_value = 원하는 값
```

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic') # 모듈에서 제공되는 데이터셋을 가져오는 매써드

df = titanic.loc[:, ['age', 'fare']] # age와 fare 칼럼의 정보만 가져옴

print(df+10)
df.add(10, fill_value=0)


df1 = df.head(10)
df2 = df.tail(10)
df1.add(df2, fill_value=0)
```

```python
age_avg = int(df.age.sum() / df.age.count())

for idx, age in enumerate(df.age):
    if age < age_avg:
        print('{} : {}'.format(df.iloc[idx,0], df.iloc[idx,1]))
```
