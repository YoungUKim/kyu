```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('./시도별 전출입 인구수.xlsx', engine='openpyxl', header=0)
df.head()

# 데이터 전처리 작업
df = df.fillna(method='ffill') # 누락 된 데이터를 앞 데이터로 채움

# 서울에서 다른 지역으로 이동한 데이터만 출력하여 정리
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
df_seoul = df_seoul.drop(['전출지별'], axis=1) # 전출지별 칼럼 삭제
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True) # 칼럼명 변경
df_seoul.set_index('전입지', inplace=True) # 칼럼명을 인덱스로 변경

df_1 = df_seoul.loc['경기도'] #x축과 y축으로 구성
plt.plot(df_1.index, df_1.values) # plt.plo(x축, y축)
plt.title("서울에서 경기도로 전출한 인구수")
plt.xlabel("연도")
plt.ylabel("인구수")
plt.xticks(rotation='vertical')
plt.legend(labels=["서울->경기"],  loc='upper right')

plt.show()
```

```python
# 문 전출지가 부산에서 다른 지역으로 전출한 데이터만 추출해서 df_busan으로 저장
# 불필요한 칼럼 제거 인덱스는 전입지별을 전입지로 바꾸어서 설정

mask2 = (df['전출지별']=='부산광역시')&(df['전입지별']!='부산광역시')
df_busan = df[mask2]
df_busan = df_busan.drop(['전출지별'], axis=1)
df_busan.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_busan.set_index('전입지', inplace=True)
df_busan.columns = df_busan.columns.map(int)

df_busan.T.plot()
```
