```python
import pandas as pd
import numpy as np

df = pd.read_csv("./auto-mpg.csv", header = None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
             'accelation','model year','origin','name']
```

```python
df['horsepower'].replace('?',np.nan, inplace=True)
df.dropna(subset=['horsepower'], axis = 0, inplace=True)
df['horsepower'] = df['horsepower'].astype('float')
```

```python
# 3개의 구간으로 나누어 범주처리

count, bin_div = np.histogram(df['horsepower'], bins=3) #경계 리스트
print(bin_div)

bin_names = ['저출력','보통출력','고출력']
```

```python
df['hp_bin'] = pd.cut(x=df['horsepower'], #데이터 배열
                     bins = bin_div,      # 경계 리스트
                     labels = bin_names,  # bin 이름
                     include_lowest = True)

df[['horsepower','hp_bin']].head()

horse_dummies = pd.get_dummies(df['hp_bin'])
pd.get_dummies(df['hp_bin'], prefix='hp')
```

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')

titanic.dropna(subset=['age'], axis=0, inplace=True)
titanic['age'] = titanic['age'].astype('int')
count, bin_div = np.histogram(titanic['age'], bins = 4)
print(bin_div)

bin_names = ['유아기','청소년기','청년','장년']

titanic['age_bin'] = pd.cut(x = titanic['age'],
                           bins = bin_div,
                           labels = bin_names,
                           include_lowest = True)

titanic[['age', 'age_bin']].head(20)
```
